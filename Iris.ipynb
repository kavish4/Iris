```python
X=df.values[:,[0,1,2,3]]
X.shape
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,random_state=123)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.scatter(range(1,11),wcss)
plt.title("The Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("WCSS")
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,random_state=123)
y_pred=kmeans.fit_predict(X)
```

```python
from mpl_toolkits import mplot3d
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(111,projection="3d")
plt.scatter(X[y_pred == 0, 0], X[y_pred == 0, 1], s = 100, label = 'Iris-setosa')
plt.scatter(X[y_pred == 1, 0], X[y_pred == 1, 1], s = 100, label = 'Iris-versicolour')
plt.scatter(X[y_pred == 2, 0], X[y_pred == 2, 1], s = 100, label = 'Iris-virginica')

plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 70,marker='*',c = 'black', label = 'Centroids')
plt.legend()
plt.show()
```
